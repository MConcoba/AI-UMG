```python
from sklearn import datasets

iris = datasets.load_iris()
print(iris)
```

```python
x = iris['data']
y = iris['target']
names = iris['target_names']
feature_names = iris['feature_names']

print(feature_names)
```

```python
print(y)
```

```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder, StandardScaler

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

enc = OneHotEncoder()
Y = enc.fit_transform(y [:, np.newaxis]).toarray()
```

```python
print(y.toarray())
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_scaled, Y, test_size=0.3, random_state=50)
```

```python
print(y_train)
```

```python
from keras.models import Sequential
from keras.layers import Dense

from keras.utils import to_categorical



model = Sequential(name='model_p1')
model.add(Dense(8, input_dim=4, activation='relu'))
model.add(Dense(3, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```

```python
fit_callback = model.fit(x_train, y_train, batch_size=5, epochs=50, validation_data=(x_test, y_test))
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
y_pred = model.predict(x_test)
frp, tpr, thresholds = roc_curve(y_test.ravel(), y_pred.ravel())
plt.plot(frp, tpr)
print(auc(frp, tpr))
```
