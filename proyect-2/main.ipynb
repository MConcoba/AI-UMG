```python
# 1) Crear una función en Python que calcule el máximo común divisor de dos números.

def mcd(a, b):
    while a != b:
        if a > b:
            a -= b
        else:
            b -= a
    return a
# Ejemplo de uso
num1 = 48
num2 = 18
resultado = mcd(num1, num2)
print(f"El máximo común divisor de {num1} y {num2} es: {resultado}")
```

```python
# 2) Crear una función en Python que calcule el mínimo común múltiplo de dos números.

def mcm(a, b):
    product = a * b if a * b >= 0 else -a * b
    return (product) // mcd(a, b)

# Ejemplo de uso
num1 = 12
num2 = 15
resultado = mcm(num1, num2)
print(f"El mínimo común múltiplo de {num1} y {num2} es: {resultado}")
    
```

```python
# 3. Basándote en datos aleatorios (los que quieras) crea un gráfico de tarta (pie chart) con Mapplotlib.

import matplotlib.pyplot as plt
import numpy as np


# Datos aleatorios
categorias = ['A', 'B', 'C', 'D']
valores = [np.random.uniform(0, 100) for _ in categorias]

print(valores)
# Crear gráfico 
plt.figure(figsize=(8, 6))
plt.pie(valores, labels=categorias, autopct='%1.1f%%', startangle=140)
plt.title('Gráfico de Tarta de Datos Aleatorios')
plt.axis('equal')  # Para que el gráfico sea un círculo
plt.show()
```

```python
# 4. Basándote en datos aleatorios, crea un Scatter Plot con Mapplotlib. Explica para qué pueden ser útiles estos gráficos.

# Datos aleatorios
x = np.random.rand(50)
y = np.random.rand(50)
# Crear gráfico
plt.figure(figsize=(8, 6))
plt.scatter(x, y, c='blue', alpha=0.5)
plt.title('Scatter Plot de Datos Aleatorios')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True)
plt.show()
# Explicación
# Los scatter plots son útiles para visualizar la relación entre dos variables.
# Pueden ayudar a identificar patrones, tendencias o correlaciones en los datos.
```

```python
# 5. Explica cómo combinar 4 gráficos en una única figura (subplot)

# Para combinar múltiples gráficos en una única figura, se puede utilizar la función `subplots` de Matplotlib.
# Esta función permite crear una cuadrícula de subgráficos en una sola figura.

# Crear datos aleatorios
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)
# Crear figura y subgráficos
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
# Graficar en cada subgráfico
axs[0, 0].plot(x, y1, 'r')
axs[0, 0].set_title('Seno')
axs[0, 1].plot(x, y2, 'g')
axs[0, 1].set_title('Coseno')
axs[1, 0].plot(x, y3, 'b')
axs[1, 0].set_title('Tangente')
axs[1, 1].plot(x, y4, 'm')
axs[1, 1].set_title('Exponencial')
# Ajustar el espacio entre subgráficos
plt.tight_layout()
# Mostrar la figura
plt.show()
```
